```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("banglore.csv")
df
```

```
df1 = df.drop(['area_type', 'availability', 'society'], axis='columns')
df1.head()
```

```
df1.shape
```

```
df1.isnull().sum()
```

```
df1 = df1.drop('balcony', axis=1)
df1.head()
```

```
df1.dropna(inplace=True)
df1.isnull().sum()
```

```
df1['size'].unique()
```

```
df1['bhk'] = df1['size'].apply(lambda x: int(x.split(' ')[0]))
df1['bhk'].unique()
```

```
df1.head()
```

```
df1.drop(['size'], axis=1, inplace=True)
```

```
df1.head()
```

```
df1['total_sqft'].unique().tolist()
```

```
def convert_number(x):
  token = x.split('-')
  if(len(token)==2):
    return (float(token[0])+float(token[1]))/2
  try:
    return float(x)
  except:
    return None
```

```
convert_number('1004 - 1204')
```

```
df2 = df1.copy()
df2['total_sqft'] = df2['total_sqft'].apply(convert_number)
df2['total_sqft']
```

```
df2['total_sqft'].unique().tolist()
```

```
df2.loc[410]
```

```
df2.dropna(inplace=True)
```

```
df2['price_per_sqft'] = (df2.price*100000)/df2.total_sqft
df2['price_per_sqft']
df2.head()
```

```
len(df2.location.unique())
```

```
location_stats = df2.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```
less_than_10 = location_stats[location_stats<=10]
```

```
df2['location'] = df2['location'].apply(lambda x: 'other' if x in less_than_10 else x)
len(df2.location.unique())
```

```
df2.shape
```

```
df2[df2.total_sqft/df2.bhk<300].head()
```

```
df3 = df2[~(df2.total_sqft/df2.bhk<300)]
```

```
df3.shape
```

```
df3.price_per_sqft.describe()
```

```
def remove_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(df.price_per_sqft)
    std = np.std(df.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft>(m-std)) & (subdf.price_per_sqft<=(m+std))]
    df_out = pd.concat([df_out, reduced_df], ignore_index=True)
  return df_out

df4 = remove_outliers(df3)
```

```
df4.shape
```

```
def remove_bhk_outliers(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean':np.mean(bhk_df.price_per_sqft),
          'std':np.std(bhk_df.price_per_sqft),
          'count':bhk_df.shape[0]
        }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')


df4 =  remove_bhk_outliers(df4)
df4.shape
```

```
df5 = df4[df4.bath<df4.bhk+2]
df5.shape
```

```
df5.drop(['price_per_sqft'],axis=1,inplace=True)
```

```
df5
```

```
dummies = pd.get_dummies(df5.location)
dummies
```

```
df6 = pd.concat([df5, dummies.drop('other', axis=1)], axis='columns')
```

```
df6.head()
```

```
df6.drop('location', axis=1, inplace=True)
```

```
X=df6.drop('price', axis=1)
X.head()
```

```
Y = df6.price
Y
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```
